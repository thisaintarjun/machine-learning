# Series v/s Dataframes

```python
'''series is like a labelled column whereas dataframes are like tables'''
import pandas as pd
print(pd.__version__)
```

# Series

```python
'''A series is a one dimensional labelled array that can hold any data type, think of it 
like a single column of a spreadsheet(1-D)'''
data=[100,102,104]
series=pd.Series(data) #This is a  constructor not a function and series is a object
print(series) 
'''we got labels on the left, values on the right and metadata on the bottom. This metadata
can be of any type, float-boolean-float etc.'''
#we can also label ourselves
series=pd.Series(data, index=['A','B','C'])
print(series)
#value of label in series
print(series.loc['C'])
#we can also change the value using loc
series.loc['C']=200
print(series)
#to locate a value by integer position
print(series.iloc[1])
#Filter by values
print(series[series>100]) 
```

# Dataframes

```python
import pandas as pd 
data={
    "Name":["Spongebob","Patrick","Squidward"],
     "Age":[30,35,50]
}
df=pd.DataFrame(data, index=["Employee_1","Employee_2","Employee_3"])
print(df)
#this is to select a row 
print(df.loc["Employee_1"])
#select a row from integer position
print(df.iloc[1])
#Add a new column 
df["Job"]=["Cook", "N/A","Cashier"]
print(df)
#to create a new row
#pass as many dictionary for as many rows you want(here 1)
new_row=pd.DataFrame([{"Name":"Sandy","Age":20,"Job":"Engineer"}],index=["Employee_4"])
df=pd.concat([df,new_row])
print(df)
#Add new rows 
new_rows=pd.DataFrame([{"Name":"Eugene","Age":40,"Job":"Manager"},
                      {"Name":"Boomer","Age":28,"Job":"Janitor"}],
                      index=["Employee_5","Employee_6"])
df=pd.concat([df,new_rows])
print(df)
```

```python
#selection by row
#print(df.loc[0])
'''since we won't remember that which pokemon is at which number, what
we can do is that mark them/ index them using pokemon's name's column'''
df=pd.read_csv("pokemon.csv", index_col="Name")
print(df.loc["Pikachu"]) #now we get data of pikachu
print(df.loc["Pikachu",["Weight",]])
'''we pass the first list as range or how many things you want data of
and the other list is the data that you want'''

#using indices
print(df.iloc[0:11:2, 0:3]) #start end step, range of column 
```

# Aggregation
Aggregate Functions: Reduces a set of values into a single summary value used to summarize and analyze data often used with the groupby() function

```python
import pandas as pd
df=pd.read_csv("Pokemon.csv")
#print(df.mean(numeric_only=True)) #function works only on numeric columns
'''similary we have the following functions ALL THESE WORKS ON 
ENTIRE DATAFRAME:
sum()
min()
max()
count()'''
'''similary we have the following functions ALL THESE WORKS ON 
SINGLE COLUMN:
mean()
sum()
similary other functions'''
print(df["Height"].mean())
```

```python
#GROUP BY
import pandas as pd
df=pd.read_csv("Pokemon.csv")
group=df.groupby("Type1")
print(group["Height"].sum())
'''what this means is that we are grouping pokemon by their type 1s and
this is done on basis of sum of their heights
these are not sorted in any way though but can be used for that
other functions can be used here in place of sum()'''
```
