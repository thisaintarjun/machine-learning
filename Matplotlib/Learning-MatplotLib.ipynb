# PLOT CUSTOMISATIONS

```python
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20]) 

#BEFORE CUSTOMIZATION
plt.plot(x, y)
plt.show()

#AFTER CUSTOMIZATION
plt.plot(x, y, marker=".", markersize=30,
        markerfacecolor="cyan", markeredgecolor="black",
        linestyle="dashed", linewidth=4, color="black")
plt.show() 

# marker is used to mark the coordinates, there are specific 
# set of markers 
# markersize or ms is used to set markersize
# markerfacecolor or mfc is used to change the color of the marker in
# which you can input hexadecimal value of specific color or just name of known 
# color
# markeredgecolor or mec is used to set the color of the marker's edge
# linestyle is used to change the style of the line into 
# dashed, dotted, dashdot, solid or just none for no line
# linewidth to change the width of the line
# color is used to change the color of the line
```

```python
# plotting two graphs together
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20]) 
y2=np.array([15,25,35,20])
#dictionary of customizations 
 # custom={"marker":".", "markersize":30,
 #        "markerfacecolor":"cyan", "markeredgecolor":"black",
 #        "linestyle":"dashed", "linewidth":4, "color":"black"}
#I changed into dictionary manually whereas I could've just used dict() function the following way
custom=dict( marker=".", markersize=30,
        markerfacecolor="cyan", markeredgecolor="black",
        linestyle="dashed", linewidth=4, color="black")
#now we can just pass this dictionary whenever we need to customise the graph by unppacking using **
#graph01
plt.plot(x, y2)
#graph02
plt.plot(x, y,**custom)
#plotting both of them together
plt.show() 
# we can also change the attribute inside of our dictionary of customization extensively after unpacking by just passing it 
# as an argument. More than two graphs can be plot on a single graph and changing their individual colors help in 
# seeing them better
```

# LABELS

```python
#Adding Titles and stuff
import matplotlib.pyplot as plt
import numpy as np 
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20]) 
y2=np.array([15,25,35,20])
#TITLE
plt.title("class size", fontsize=25, family="Arial",
         fontweight="bold", color="black")

#LABEL FOR X-AXIS
plt.xlabel("Year",fontsize=20,family="Arial",
          fontweight="bold",color="blue")

#LABEL FOR Y-AXIS
plt.ylabel("Students",fontsize=20,family="Arial",
          fontweight="bold",color="blue")

#CUSOMISATION FOR TICKS to only come in certain places FOR X-AXIS
# plt.xticks(x)
#to customise the ticks
plt.tick_params(axis="both",colors="green")


custom=dict( marker=".", markersize=30,
        markerfacecolor="cyan", markeredgecolor="black",
        linestyle="dashed", linewidth=4, color="black")
plt.plot(x, y2)
plt.plot(x, y,**custom)
plt.show() 
```

# BAR CHARTS
compare categories of data by representing each category with a bar

```python
import matplotlib.pyplot as plt
import numpy as np 
categories=np.array(["grains","Fruit","Vegetables","Protein","Dairy","Sweets"])
values=np.array([4,3,2,5,3,1])

plt.bar(categories, values,color="black")
#however, if we want a horizontal bar chart:-
# plt.barh(categories, values,color="black")
plt.title("Daily consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()

```

# PIE CHARTS
A circular chart divided into slices to show percentages of the total.
Good for visualizing distribution among categories.

```python
import matplotlib.pyplot as plt
import numpy as np 
categories=np.array(["Freshman","Sophomores","Juniors","Seniors"])
values=np.array([300,250,275,225])
colors=["red","yellow","blue","green"]
plt.pie(values,labels=categories,
               autopct="%1.1f%%"
               ,colors=colors
               ,explode=[0,0,0,0.1],
               shadow=True
               ,startangle=90) 
#formatting instruction to add percentages in the pie-chart
#we use double percent sign to show % in the pie chart because using one
# means its a format specifier
#in the explode the more the amount the farther the distance
#we can use shadows and we can also rotate using startangle and passing 
#degrees with want to rotate with
plt.title("College")
plt.show()
```

# SCATTER PLOTS
Shows the relationship between two variables helps to identify a correlation(+,-, none)
Example: Study hours v/s Test scores

```python
import matplotlib.pyplot as plt
import numpy as np
x1=[0,1,1,2,3,4,5,6,7,7,8] #hours studied
y1=[55,60,65,62,68,70,75,78,82,85,87]#test scores
x2=[0,1,2,2,3,4,5,6,7,8,8] 
y2=[50,58,65,70,72,78,83,88,92,95,97]
plt.scatter(x1,y1,color="skyblue",alpha=0.5,s=200,label="Class-A")
plt.scatter(x2,y2,color="red",alpha=0.5,s=200,label="Class-B")
#We need a legend here to differentiate between both the plots
#before that we need to individually label both the plots as done above
plt.legend()
#the color is of the dot that we changed above
# the alpha refers to the opacity of the dots which we set to 0.5 here
# and the s is just an abbreviation for size of the dots 
plt.title("TEST SCORES")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.show()
```

# HISTOGRAMS
A Visual representation of the distribution of quantitative data.
They group values into bins(intervals) and counts how many falls in each range

```python
import matplotlib.pyplot as plt
import numpy as np
#we are creating a numpy array of random scores
# loc means the location of the center(median of our data)
# all of the test scores will be around loc that is passing scores
# scale is the daviation from loc the higher the input
# the greater the deviation
# size means we have 100 students' grades in our array

scores=np.random.normal(loc=80,scale=10,size=100)
#to ensure that there are no scores below 0 and above 100 we use clip()
scores=np.clip(scores,0,100)#setting min and max here
plt.hist(scores,bins=10,color="lightgreen", #we set the number of bins 
         #if we want to
        edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("Scores")
plt.ylabel("# of students")
plt.show()
```

# SUBPLOTS
 Figure is the entire campus and Ax is a single plot(subplot)

```python
import matplotlib.pyplot as plt
import numpy as np

print(plt.subplots(2,2))#2 rows and 2 columns 
#subplot function returns a tuple. A tuple of 2 entities a figure object
# which is our canvas and a 2-d array of Axes objects
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])

figure, axes=plt.subplots(2,2)
#we fill them individually
axes[0,0].plot(x,x*2,color="red")
axes[0,0].set_title("x*2")


axes[0,1].plot(x,x**2,color="blue")
axes[0,1].set_title("x**2")


axes[1,0].plot(x,x**3,color="green")
axes[1,0].set_title("x**3")


axes[1,1].plot(x,x**4,color="purple")
axes[1,1].set_title("x**4")

#Here we used plots but we can also use other graphs for example bar 
# charts, pie charts, histogram
# and it is not necessary to use the same one for all we can use different
# for all or any of the sublots
plt.show()
```

# PANDAS + MATPLOTLIB
intgrating pandas and matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
df=pd.read_csv("Pokemon.csv")
# print(df["Type1"]) #Selecting single column to use here
#value_counts give us the number of all the types1
type_count=df["Type1"].value_counts(ascending=True) #type_count will be a series
plt.barh(type_count.index, type_count.values,color="skyblue",edgecolor="black")
plt.title("# of Pokemon by primary type")
plt.xlabel("Count")
plt.ylabel("Type")
plt.tight_layout()#to make sure everything fits in our 
plt.show()
```
