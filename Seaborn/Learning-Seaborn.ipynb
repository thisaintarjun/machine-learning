```python
import seaborn as sns
# tips dataset: how much people are tipping in the restaurant and related
sns.load_dataset("tips")
# we see in our dataframe that we got day named feature so we are going to use it 
# to differentiate on that basis but using relplot
tips_data=sns.load_dataset("tips")
sns.relplot(tips_data,x="total_bill",y="tip",hue="sex",size="size",
                style="smoker",col="day",kind="scatter",col_wrap=2)  
#col_wrap is used to create a 2x2 type grid here
```

# SCATTERPLOT()
one of the major Seabor's strength is working along with hue, style and size attribute.

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# hue lets use compare groups
# style changes marker style 
# size is based on size feature of the data and shows dining party size
data=sns.load_dataset("tips")
sns.scatterplot(data,x="total_bill",y="tip",hue="sex",style="smoker"
               ,size="size")
plt.show()
```

# LINEPLOT()
shows trends

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# here we can see that graph has a positive trend
# the bigger the size of the dinner party the larger the tips
data=sns.load_dataset("tips")
sns.lineplot(data,x="size",y="tip")
plt.show()
```

# BARPLOT()
shows the average of a numeric variable

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# here we can observe the tips give on the following days
data=sns.load_dataset("tips")
sns.barplot(data,x="day",y="tip")
plt.show()
```

# COUNTPLOT()
Counts occurances

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# here we can see the count of the customers on respective days
data=sns.load_dataset("tips")
sns.countplot(data,x="day")
plt.show()
```

# HISTPLOT()
shows distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# here we can see the distribution of the bill 
data=sns.load_dataset("tips")
sns.histplot(data,x="total_bill")
plt.show()
```

# BOXPLOT()
**Excellent for finding outliers**

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# the circles denotes the outliers whereas the box represent the median
data=sns.load_dataset("tips")
sns.boxplot(data,x="day",y="total_bill")
plt.show()
```

# VIOLINPLOT()
Like a boxplot, but also shows density.

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# wider selection= more observations.
data=sns.load_dataset("tips")
sns.violinplot(data,x="day", y="total_bill")
plt.show()
```

# HEATMAP()
one of the most important plot in seaborn that is used in ML.

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# can be studied in detail later from observation till now points
# concluded are that corr, annot and cmap are used just like below used
#high positive values indicarte variables that increase toghether. 
tips=sns.load_dataset("tips")
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.show()
```

# PAIRPLOT()
Creates scatterplots for every pair of numerical column

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# seaborn makes all possible pairs of numerical automatically.
data=sns.load_dataset("tips")
sns.pairplot(data,hue="sex")
plt.show()
```

# RELPLOT()
A figure-level version of scatter/line plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# it's especially usefull when you want to split the plots into multiple
# panels using col= or row=
data=sns.load_dataset("tips")
sns.relplot(data,x="total_bill",y="tip",hue="sex"
            ,style="smoker",size="size")
plt.show()
```

# CATPLOT()
A figure-level interface for categorical plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
# it's especially usefull when you want to split the plots into multiple
# panels using col= or row=
data=sns.load_dataset("tips")
sns.catplot(data,x="total_bill",y="tip",kind="bar")
plt.show()
# below are the different types of "kind" attributes that we can use:
# kind="bar"
# kind="box"
# kind="violin"
# kind="strip"
# kind="swarm"
# kind="point"
# kind="count"
```

# PALETTE
changes colors

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
data=sns.load_dataset("tips")
sns.scatterplot(data,x="total_bill",y="tip",hue="sex",style="smoker"
               ,size="size",palette="Set1")
# The palette is used to change the colors of the 
# make sure to correctly input the name of the palette cause it is case
# sensitive
# The following are some other palettes:
# palette="deep"
# palette="Set1"
# palette="Set2"
# palette="viridis"
# palette="coolwarm"
plt.show()
```

# Figure Size
Larger values create bigger plots

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
#here to increase the size of our figure we are going to use figsize
data=sns.load_dataset("tips")
plt.figure(figsize=(10,6))
sns.scatterplot(data,x="total_bill",y="tip",hue="sex",style="smoker"
               ,size="size",palette="Set1")

plt.show()
```

# Titles and Labels
they are good for readability

```python
import seaborn as sns
import matplotlib.pyplot as plt # to use show
data=sns.load_dataset("tips")
sns.scatterplot(data,x="total_bill",y="tip",hue="sex",style="smoker"
               ,size="size",palette="Set1")
plt.title("Bill v/s Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
